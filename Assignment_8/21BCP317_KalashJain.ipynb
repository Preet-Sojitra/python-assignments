***Question : Implement all the OOPs examples provided in the PPT.***

```python
class Dog:
  def __init__(self, name, color, ear_length, spot):
    self.name = name
    self.skinColor = color
    self.earLength = ear_length
    self.isSpotted = spot
    print("Dog has been created.")
  
  def walk(self):
    print(f"{self.name} is walking!")

  def eat(self):
    print(f"{self.name} is eating!")

dog1 = Dog('Scooby', 'brown', 'short', True)
dog1.walk()

print("\n")

dog2 = Dog('Tommy', 'black', 'short', False)
dog2.walk()
```

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def myFunc(self):
    print(f"My name is {self.name}.")

person1 = Person("John", 36)
print(person1.name)
print(person1.age)
person1.myFunc()
```

***Question : Write a python code using the concept of OOPs to add complex numbers.***

```python
class Complex:
  def __init__(self, num):
    self.real = int(input(f"Real part of the {num} complex number: "))
    self.imaginary = int(input(f"Imaginary part of the {num} complex number: "))

complex1 = Complex("first")
print("\n")
complex2 = Complex("second")

print(f"\nThe sum of the complex numbers is {complex1.real + complex2.real} + {complex1.imaginary + complex2.imaginary}i.")
```
