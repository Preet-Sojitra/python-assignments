```python
principal=1000
intrest_rate=0.05
year=20
value=[]
for i in range (year+1):
  value.append(principal)
  principal+=principal*intrest_rate
pylab.plot(value)
pylab.title("5% growth, compund annualy")
pylab.xlabel("years of componding")
pylab.ylabel("Value of principal ($)")
```
